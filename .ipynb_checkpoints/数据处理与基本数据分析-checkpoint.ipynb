```python
import pandas as pd
from pandas import DataFrame
animation = pd.read_csv('output1.csv')
del animation['Unnamed: 0']
animation.head()
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 统计8分以上的作品中导演出现次数的排名(分数加权)
import numpy as np
from matplotlib.font_manager import FontProperties
import operator
font = {'family' : 'SimHei',
    'weight' : 'bold',
    'size'  : '16'}
plt.rc('font', **font)        # 步骤一（设置字体的更多属性）
plt.rc('axes', unicode_minus=False) # 步骤二（解决坐标轴负数的负号显示问题）

# 8分以上dataframe
animation_good = animation[animation['评分'] >= 8]
# 最大-最小值标准化
def MaxMinNormalization(x, min, max):
    """[0,1] normaliaztion"""
    x = (x - min) / (max - min)
    return x
# 对rank进行标准化
rank_list = list(animation_good['评分'])
rank_list = [MaxMinNormalization(x, 8.0, 9.1) for x in rank_list]
director_list = []
# 导演
for director, rank in zip(animation_good['导演'], rank_list):
    if not pd.isnull(director):
        directors = director.split(',')
        for d in directors:
            director_list.append((d, rank))
# 总导演
for director, rank in zip(animation_good['总导演'], rank_list):
    if not pd.isnull(director):
        directors = director.split(',')
        for d in directors:
            director_list.append((d, rank))
# 副导演
for director, rank in zip(animation_good['副导演'], rank_list):
    if not pd.isnull(director):
        directors = director.split(',')
        for d in directors:
            director_list.append((d, rank))

director_dict = {}
for director, number in director_list:
    director_dict[director] = director_dict.setdefault(director, 0) + number

director_dict = dict(sorted(director_dict.items(),key = operator.itemgetter(1),reverse = True))
director_dict = pd.Series(director_dict)[:10]
director_dict.plot(kind = 'bar', figsize=(16,9), use_index=True)
```

```python
# 统计8分以上作品中系列构成出现次数的排名(分数加权)
# 系列构成
script_list=[]
for script, rank in zip(animation_good['系列构成'], rank_list):
    if not pd.isnull(script):
        scripts = script.split(',')
        for s in scripts:
            script_list.append((s, rank))

script_dict = {}
for script, number in script_list:
    script_dict[script] = script_dict.setdefault(script, 0) + number

script_dict = dict(sorted(script_dict.items(),key = operator.itemgetter(1),reverse = True))
script_dict = pd.Series(script_dict)[:10]
script_dict.plot(kind = 'bar', figsize=(16,9), use_index=True)
```

```python
# 统计各个公司的作品数量排名(分数加权)
script_list=[]
for script, rank in zip(animation_good['制作公司'], rank_list):
    if not pd.isnull(script):
        scripts = script.split(',')
        for s in scripts:
            script_list.append((s, rank))

script_dict = {}
for script, number in script_list:
    script_dict[script] = script_dict.setdefault(script, 0) + number

script_dict = dict(sorted(script_dict.items(),key = operator.itemgetter(1),reverse = True))
script_dict = pd.Series(script_dict)[:10]
script_dict.plot(kind = 'bar', figsize=(16,9), use_index=True)
```

```python
# 统计各个画师的作品数量排名(分数加权)
script_list=[]
for script, rank in zip(animation_good['分镜'], rank_list):
    if not pd.isnull(script):
        scripts = script.split(',')
        for s in scripts:
            script_list.append((s, rank))

for script, rank in zip(animation_good['作画监督'], rank_list):
    if not pd.isnull(script):
        scripts = script.split(',')
        for s in scripts:
            script_list.append((s, rank))

for script, rank in zip(animation_good['原画'], rank_list):
    if not pd.isnull(script):
        scripts = script.split(',')
        for s in scripts:
            script_list.append((s, rank))
script_dict = {}
for script, number in script_list:
    script_dict[script] = script_dict.setdefault(script, 0) + number

script_dict = dict(sorted(script_dict.items(),key = operator.itemgetter(1),reverse = True))
script_dict = pd.Series(script_dict)[:10]
script_dict.plot(kind = 'bar', figsize=(16,9), use_index=True)
```

```python
# 统计各个公司的作品的平均分
rank_list = list(animation['评分'])
script_list=[]
for script, rank in zip(animation['制作公司'], rank_list):
    if not pd.isnull(script):
        scripts = script.split(',')
        for s in scripts:
            script_list.append((s, rank))
script_list = list([list(x) for x in script_list])
script_list = pd.DataFrame(script_list, columns=['company', 'rank'])

company_set = set(script_list['company'])
for company in company_set:
    if script_list[script_list['company']==company]['company'].count() < 5:
        script_list=script_list[~script_list['company'].isin([company])]

groups_rank_company = script_list["rank"].groupby(script_list["company"])
mean_rank = groups_rank_company.mean()
mean_rank.index.name = ""
mean_rank = mean_rank.sort_values(ascending=False)[:10]
mean_rank.plot(kind = 'bar', figsize=(16,9), use_index=True)
```

```python
# 看过人数-分数关系
sns.jointplot(x='看过',y='评分',data=animation)
plt.show()
```

```python
# 总集数-弃坑/搁置人数的关系
animation = animation.drop(animation[animation['总集数'] > 1000].index)
sns.jointplot(x='总集数',y='弃坑',data=animation)
plt.show()
```
